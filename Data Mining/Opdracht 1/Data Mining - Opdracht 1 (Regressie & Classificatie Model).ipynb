Het Random Forest regressiemodel wordt in de praktijk op veel verschillende manieren toegepast. Er zijn vooral  veel voorbeelden te vinden van economische onderzoeksmodellen. Zo is er een onderzoek uit 2021 dat Random Forest inzet om de impact van inflatie op economische groei te analyseren. 

In het rapport *"Inflation and Inflation Uncertainty in Growth Model of Barro: An Application of Random Forest Method"* wordt Random Forest gebruikt om te onderzoeken in hoeverre zowel hoge als lage inflatie samenhangen met tragere economische groei. Het model helpt  om voorspellingen te doen, maar ook om andere factoren te identificeren die van invloed zijn op economische groei en variabelen te sorteren op basis van hun belang voor de voorspelling. 

2. Hoe werkt het algoritme conceptueel? Wat zijn de belangrijkste stappen?

Een Random Forest regressiemodel bestaat uit een verzameling beslissingsbomen die elk worden getraind op een willekeurige subset van de trainingsdata én op een willekeurige subset van de beschikbare features. Elke boom doet een eigen voorspelling, en de uiteindelijke voorspelling van het Random Forest is het gemiddelde van alle individuele voorspellingen. 

Omdat elke boom is opgebouwd met een andere subset van de trainingsdata en een willekeurige subset van voorspellende variabelen, hebben random forests ingebouwde variabiliteit. Deze variabiliteit helpt overfitting te voorkomen. 

De nauwkeurigheid van het model wordt bepaald aan de hand van:
* R-squared (r2_score): hoeveel van de spreiding in de target (SalePrice) wordt verklaard door het model. → **Hoe hoger, hoe beter**.
* Mean Squared Error (mse): de gemiddelde kwadratische fout tussen de echte en voorspelde waarden. → **Hoe lager, hoe beter**.
* OOB Score (oob_score_): Out-of-Bag” Score is een soort ingebouwde cross-validation binnen Random Forest. → **Hoe hoger, hoe beter**.


De belangrijkste stappen zijn: 
1. Datavoorbereiding - (missende waarden behandelen, encoderen, schalen indien nodig)
2. Data splitsen - in een trainings- en testset (of gebruikmaken van OOB-validatie)
3. Model trainen - op de trainingsdata
4. Model evalueren - op testdata of via OOB-score
5. Resultaten visualiseren - (bijv. voorspelde vs. werkelijke waarden, feature importance)

3. Wat zijn de voor- en nadelen van jullie algoritme? In welke situaties werkt het heel goed en wanneer juist niet?

Voordelen van Random Forest:
* Hoge nauwkeurigheid van de voorspellingen
* Geschikt voor zowel continue variabelen als categorische variabelen
* Minder overfitting
* Kan omgaan met niet-lineare relaties tussen target en features
* Efficiënt bij grote datasets

Nadelen van Random Forest:
* Het model kan moeilijk te interpreteren zijn
* Trager bij voorspelen en trainen dan eenvoudigere modellen
* Groot geheugengebruik
* Niet snel genoeg voor real-time systemen
---

Geschikte inzet van Random Forest:
- Bij grote en complexe datasets met veel variabelen
- Wanneer nauwkeurigheid belangrijk is 
- Bij datsets met veel ruis of outliers

Niet geschikte inzet van random forest:
- Bij kleine datasets
- Wanneer interpretatie en uitlegbaarheid belangrijk is 
- Bij real-time systemen

### Lineaire Regressie model

Maak een enkelvoudig lineaire regressie model (dus met maar één X-variabele, en NIET jouw gekozen regressie model...dit is bedoeld als een soort inleiding) net als de demo in het college.

```python
# Enkelvoudig lineaire regressie model 
# x, y instellen
X = df_train[['OverallQual']]
y = df_train['SalePrice']

# model maken
model = LinearRegression()
model.fit(X, y)

# Voorspellingen maken
y_pred = model.predict(X)

# Plots maken Scatter + Regressielijn
plt.figure(figsize=(8, 6)) # Formaat
plt.scatter(X, y, alpha=0.5, label='Data') # Scatterplot
plt.plot(X, y_pred, color='red', linewidth=2, label='Regressielijn') #Regressielijn
plt.title("Lineaire Regressie: SalePrice vs OverallQual")
plt.xlabel("OverallQual")
plt.ylabel("SalePrice")
plt.legend()
plt.tight_layout()
plt.show()

# Print modelcoëfficiënten
intercept = model.intercept_
coef = model.coef_[0]
print(f"Regressievergelijking: SalePrice = {intercept:.2f} + {coef:.2f} * OverallQual")
```

Wat is de vergelijking die hierbij hoort?

De vergelijking van de regressielijn is SalePrice = -96206.08 + 45435.80 * OverallQual

### Residuen Plot

Maak een residuen plot (de input variabele op de x-as en de fout in de voorspelling op de y-as). Zie voorbeeld hier rechtsonder.

```python
# Residuen plot maken
# Residuen berekenen
residuen = y - y_pred  # y = echte waarde, y_pred = voorspelde waarde

# Plotten
plt.figure(figsize=(8, 6)) 
plt.scatter(X, residuen, alpha=0.5)
plt.axhline(0, color='red', linestyle='--', linewidth=1)
plt.title("Residuenplot: SalePrice vs OverallQual")
plt.xlabel("OverallQual")
plt.ylabel("Residuen (fout in voorspelling)")
plt.tight_layout()
plt.show()
```

Wat valt je op?

Het valt op dat bij een lagere (0-4) OverallQual de fouten meestal positief zijn, en het model de verkoopprijs dus te laag inschat. Vervolgens is het bij een gemiddelde (5-8) OverallQual vaker dat de verkoopprijs te laag word ingeschat. En bij een hoge (9-10) OverallQual is een stuk grotere spreiding te zien dan bij de lagere OverallQual.  

### Gekozen Regressie Model

Maak nu een voorspelling met jullie gekozen machine learning model en gebruik eerst dezelfde enkele variabele van de enkelvoudig lineaire regressie jullie net hebben gedaan.

```python
# Regressiemodel met één variabele
# x, y instellen
X = df_train[['OverallQual']]
y = df_train['SalePrice']

# Model aanmaken (met OOB)
regressor = RandomForestRegressor(
    n_estimators=100, # Gebruik 100 bomen
    oob_score=True, # out-of-bag validatie aan
    random_state=42, # seed instellen
    bootstrap=True) 

# Model trainen
regressor.fit(X, y)

# Voorspellingen maken
y_pred = regressor.predict(X)

# Model evalueren
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
oob_score = regressor.oob_score_

# printen scores
print(f"Random Forest MSE: {mse:.2f}")
print(f"Random Forest R²: {r2:.4f}")
print(f"Out-of-Bag Score (OOB): {oob_score:.4f}")

# Scatterplot maken w vs v
plt.figure(figsize=(8, 6))
plt.scatter(X, y, alpha=0.5, label='Werkelijke Waarden')
plt.scatter(X, y_pred, color='green', alpha=0.4, label='Voorspellingen (RF)')
plt.title("Random Forest Regressie: SalePrice vs OverallQual")
plt.xlabel("OverallQual")
plt.ylabel("SalePrice")
plt.legend()
plt.tight_layout()
plt.show()

# Plot van een enkele boom uit de Random Forest
tree_to_plot = regressor.estimators_[0]  # Eerste boom in het bos

# Plot de decision tree
plt.figure(figsize=(20, 10))
plot_tree(
    tree_to_plot,
    feature_names=X.columns.tolist(),  # Gebruik alleen de gebruikte features
    filled=True,
    rounded=True,
    fontsize=10)

plt.title("Decision Tree uit Random Forest")
plt.show()
```

```python
# Regressiemodel met sterk correlerende variabelen
# Correlatie berekenen
cor_matrix = df_train.corr(numeric_only=True)
cor_saleprice = cor_matrix['SalePrice'].drop('SalePrice')

# Selecteer variabelen met 'Zeer sterke' correlatie (|r| > 0.7)
zeer_sterk_features = cor_saleprice[cor_saleprice.abs() > 0.7].index.tolist()
print(f"Zeer sterk correlerende variabelen: {zeer_sterk_features}")

#  Data voorbereiden
X_strong = df_train[zeer_sterk_features]
y = df_train['SalePrice']

# Model aanmaken 
regressor_strong = RandomForestRegressor(
    n_estimators=100,
    oob_score=True,
    random_state=42,
    bootstrap=True)

# Model trainen
regressor_strong.fit(X_strong, y)

# Voorspellingen maken
y_pred_strong = regressor_strong.predict(X_strong)

# Model evaluatie
mse_strong = mean_squared_error(y, y_pred_strong)
r2_strong = r2_score(y, y_pred_strong)
oob_score_strong = regressor_strong.oob_score_

# Printen features en scores
print(f"\n🔍 Model met alleen 'Zeer sterk' correlerende variabelen:")
print(f"Features: {zeer_sterk_features}")
print(f"Random Forest MSE: {mse_strong:.2f}")
print(f"Random Forest R²: {r2_strong:.4f}")
print(f"Out-of-Bag Score (OOB): {oob_score_strong:.4f}")

# 6. Plot 
plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred_strong, alpha=0.5, color='purple')
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--', color='gray')
plt.title("Voorspelling vs Werkelijke SalePrice (Zeer sterke features)")
plt.xlabel("Werkelijke SalePrice")
plt.ylabel("Voorspelde SalePrice")
plt.tight_layout()
plt.show()
```
